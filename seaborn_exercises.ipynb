```python
# displot for total_bill (hist, kde, rug)

sns.displot(data = iris, x = 'petal_length', kde = True, rug= True)
```

```python
sns.relplot(data = iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
```

```python
plt.figure(figsize=(12, 10))

plt.subplot(221)
sns.boxplot(y = 'sepal_length', x = 'species', data = iris)
plt.title('Sepal Length')

plt.subplot(222)
sns.boxplot(y = 'sepal_width', x = 'species', data = iris)
plt.title('Sepal Width')

plt.subplot(223)
sns.boxplot(y = 'petal_length', x = 'species', data = iris)
plt.title('Petal Length')

plt.subplot(224)
sns.boxplot(y = 'petal_width', x = 'species', data = iris)
plt.title('Petal Width')
```
